```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
import plotly.io as pio
pio.renderers.default='colab'#to initialize plotly
```

```python
df=pd.read_csv('/content/covid_worldometer_data.csv')
```

```python
df
```

```python
px.bar(df.head(15),x='Country',y='TotalCases',color='Country')#every country has its own color
```

```python
px.bar(df.head(15),x='Country',y='TotalDeaths',color='Country')
```

```python
df.columns
```

```python
px.scatter(df,x='Totalcases/1M',y='TotalDeath/1M',color='Country')
```

```python
#a choropleth map displays divided geographical areas or regions that are coloured,shaded or patterned in
#relation
```

```python
df.columns
```

```python
px.choropleth(df,locations='iso_alpha',hover_name='Country',color='TotalDeath/1M',
              projection='orthographic',color_continuous_scale='Viridis')
```

```python
px.choropleth(df,locations='iso_alpha',hover_name='Country',color='TotalDeath/1M',
              projection='equirectangular',color_continuous_scale='plasma')
```

```python
px.choropleth(df,locations='iso_alpha',hover_name='Country',color='TotalDeath/1M',
              scope='asia',color_continuous_scale='Viridis')
```

```python
filter1=df1[(df1['State']=='US')]
```

```python
px.bar(df1,x='Age Group',y='COVID-19 Deaths',color='Condition',hover_name='Condition',log_y=True)
```

```python
px.bar(filter1,x='Age Group',y='COVID-19 Deaths',color='Condition Group',hover_name='Condition',)
```

```python
filter2=df1[(df1['State']=='US')&(df1['Age Group']=='All Ages')]
```

```python
filter2.columns
```

```python
px.bar(filter2,x='Condition Group',y='COVID-19 Deaths',color='COVID-19 Deaths',hover_name='Condition')
```

```python
filter3=df1[(df1['Condition']=='Influenza and pneumonia') & (df1['Age Group']=='All Ages')]
```

```python
filter3
```

```python
px.bar(filter3,x='State',y='COVID-19 Deaths',color='State',log_y=True)
```

```python
#convert the columns with condition group into list using tolist() function
#convert the list to one single string
#feed the string to WordCloud
#WordCloud vis. need matplotlib
```

```python
list1=filter2['Condition Group'].tolist()
```

```python
string1=''.join(list1)
```

```python
plt.figure(figsize=(20,20))
plt.imshow(WordCloud().generate(string1))
```
